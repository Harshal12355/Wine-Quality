# Preproccesing

```python
new_quality = []

for i in df['quality']:
    if i >= 5:
        new_quality.append(1)
    else:
        new_quality.append(0)
```

```python
new_quality
```

```python
new_df = df.copy()
```

```python
new_df['new_quality'] = pd.DataFrame(new_quality)
```

```python
new_df
```

```python
new_df = new_df.drop(columns=['quality'], axis=1)
```

```python
new_df.corr()['new_quality'].sort_values(ascending=False)
```
